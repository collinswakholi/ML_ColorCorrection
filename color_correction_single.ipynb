```python
# read image
img = cv2.imread('Image/Berries/DSC_0331image32.png')

# check if image is uint8 else convert
if img.dtype != np.uint8:
    img = img.astype(np.uint8)
```

```python
# create instance of ColorCorrection class
cc = ColorCorrection(img)

# extract color chart
# src, img_draw = cc.extract_color_chart()
# run color correction
corr_img, img_draw, dims = cc.run_correction(resize_f=1, n_components=15, degree=3, show=True)

print('patch dims: ', dims)

# write raw image and color corrected image to disk
# cv2.imwrite('DSC_0331image32_raw.png', img)
# cv2.imwrite('DSC_0331image32_corr.png', corr_img)

```
